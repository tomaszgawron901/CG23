0. Any 3 point can connected so they create a polygon

1. we can take 3 most left points and create a triangle of them. There is no other point inside this triangle. Remove left most point and repeat the procedure.
The invariant is that after every iteration on the left we have triangles that every one of them has one common edge with one other and they doesn't overlap, to the right we have points that are not yet triangulated.
At the end of the procedure we only have triangles that are connected and not overlap each other - those triangles create a simple polygon
